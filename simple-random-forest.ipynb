```python

import numpy as np 
import pandas as pd 
import os
from sklearn.ensemble import RandomForestClassifier
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
```

```python
train_data = train_data[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
# Convert 'Sex' to numeric: male = 0, female = 1
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
train_data = train_data.dropna()

# Calculate correlation with 'Survived'
correlation_with_survived = train_data.corr()['Survived'].sort_values(ascending=False)
print("Correlation with Survived:")
print(correlation_with_survived)
```

```python
features = ["Pclass", "Sex", "Parch"]

# Use get_dummies on both sets
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])
X, X_test = X.align(X_test, join='left', axis=1, fill_value=0)
# Fit and predict
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, train_data['Survived'])
predictions = model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
